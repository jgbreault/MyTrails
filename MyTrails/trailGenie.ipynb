```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
%matplotlib inline
np.random.seed(42)
```

```python
allTrailsFormattedData = pd.read_csv("allTrailsFormattedData.csv", encoding="iso-8859-1")
jamieHikesCompleted = pd.read_csv("jamieHikesCompleted.csv", encoding="iso-8859-1")
trails = pd.concat([allTrailsFormattedData, jamieHikesCompleted])
trails = trails.sample(frac=1) #Shuffle data
print(trails[trails.rating == 0].count()[0]/len(trails))  #Monkey
X = trails.drop("id", axis=1).drop("name", axis=1).drop("rating", axis=1)
y = trails.rating


catagorical_features = ["route_type"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot", one_hot, catagorical_features)], remainder="passthrough")
X = transformer.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_c = jamieHikesCompleted.drop("id", axis=1).drop("name", axis=1).drop("rating", axis=1)
y_c = jamieHikesCompleted.rating
X_c = transformer.fit_transform(X_c)

jamieHikesWatchList = pd.read_csv("jamieHikesWatchList.csv", encoding="iso-8859-1")
X_w = jamieHikesWatchList.drop("id", axis=1).drop("name", axis=1).drop("rating", axis=1)
X_w = transformer.fit_transform(X_w)
```

```python
grid = {"n_estimators": [100, 200, 500, 1000],
        "max_depth": [5, 10, 20, 30],
        "min_samples_split": [2, 4, 6],
        "min_samples_leaf": [1, 2, 4]}

model = RandomForestClassifier()
model = RandomizedSearchCV(estimator=model, param_distributions=grid, n_iter=144, cv=5, verbose=10)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```
